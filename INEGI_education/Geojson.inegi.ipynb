```python
import pandas as pd
import geopandas as gp
```

```python
datos_censo = pd.read_csv('INEGI_data/iter_00_cpv2010/conjunto_de_datos/iter_00_cpv2010.csv')
```

```python
datos_censo.head(10)
```

```python
datos_municipales = datos_censo[datos_censo.nom_loc == "Total del Municipio"]
datos_municipales.head()
```

```python
datos_municipales["CVE_ENT"] = datos_municipales.entidad.astype(str).str.zfill(2)
datos_municipales["CVE_MUN"] = datos_municipales.mun.astype(str).str.zfill(3)
```

```python
datos_municipales.head(2)
```

```python
poligonos = gp.read_file("INEGI_data/889463142683_s/conjunto_de_datos/areas_geoestadisticas_municipales.shp")
```

```python
poligonos.head()
```

```python
# municipios_educacion = datos_municipales[["CVE_ENT", "CVE_MUN", "graproes"]].merge(poligonos, on = ["CVE_ENT", "CVE_MUN"], how = "right")
municipios_educacion = poligonos.merge(datos_municipales[["CVE_ENT", "CVE_MUN", "graproes"]], on = ["CVE_ENT", "CVE_MUN"], how = "left")
```

```python
municipios_educacion.head(3)
```

```python
municipios_educacion = municipios_educacion.to_crs("EPSG:4326")
municipios_educacion.head()
```

```python
municipios_educacion["geometry"] = municipios_educacion["geometry"].simplify(tolerance = 0.001, preserve_topology = True)
```

```python
municipios_educacion.to_file("INEGI_data/municipios_educacion.geojson", driver='GeoJSON')
```
